# Analisis de datos

## Formulas a utilizar

* Datos convertidos en VoltajeM
$Datos[V]=(\frac{ValorDato \ast VDIV}{25})-OFST$

* Distancia total del evento
$NroTotal=SARA \ast TDIV \ast 14 $

* Distancia temporal de cada intervalo de eventos
$H.V =2.00E-05 [s]$

* Tiempo Total del Osciloscopio
$TiempoTotal=TDIV \ast 14$

## Lectura de Archivo

```python
#importamos librerias a utilizar
import os
import re
import pandas as pd
import numpy
```

```python
# lee el contenido del archio de texto donde se esta almacenado
# vectores de almacenamiento
event_data = event = []
contador_evento = contador_suceso = 0
# Character control for identifying headers
caracter_control = "#"
# Capture of empty field values
VDIV = OFST = TDIV = SARA = 0.0
# Capture of text field values
C_VDIV = C_OFST = C_TDIV = C_SARA = ""
# Text values to search in the string
T_VDIV = "VDIV"
T_OFST = "OFST"
T_TDIV = "TDIV"
T_SARA = "SARA"
#manejo de fechas
start_date = end_date = None
T_DATE = r"(Mon|Tue|Wed|Thu|Fri|Sat|Sun) (\w{3}) (\d{1,2}) (\d{2}:\d{2}:\d{2}) ([\+\-]\d{4}) ([\+\-]\d{2}) (\d{4})"
```

```python
with open('../documentos/DAQ_20K_05.07.22.txt', 'r', encoding='latin1') as file:
    for line in file:
        if line.startswith(caracter_control):
            if "Tue" in line or "Mon" in line or "Wed" in line or "Thu" in line or "Fri" in line or "Sat" in line or "Sun" in line:
                if not start_date:
                    start_date=line.strip().split('#')[1]                    
                else:
                    end_date=line.strip().split('#')[1]                    
            contador_evento += 1
            event = []

            # Check if the line contains all required text values
            if all(text in line for text in [T_VDIV, T_OFST, T_TDIV, T_SARA]):
                cadena = line.split(' ')
                # Separate text data from the header
                for i, word in enumerate(cadena):
                    if word.endswith(T_VDIV):
                        C_VDIV = float(cadena[i + 1].split('V')[0])
                    elif word.endswith(T_OFST):
                        C_OFST = float(cadena[i + 1].split('V')[0])
                    elif word.endswith(T_TDIV):
                        C_TDIV = float(cadena[i + 1].split('S')[0])
                    elif word.endswith(T_SARA):
                        C_SARA = float(cadena[i + 1].split('Sa/s')[0])
        elif re.match(re.compile(r'^\-?[1-9][0-9]*$'), line):
            event.append(int(line))
        elif len(line.strip()) == 0 and contador_suceso == 0:
            total_event = C_SARA * C_TDIV * 14
            if len(event) != int(total_event):
                event += [0] * (int(total_event) - len(event))
            else:
                print(len(line))
            contador_suceso += 1
            event_data.append(event)
        else:
            contador_suceso = 0  # Comment explaining the purpose of this line
```
